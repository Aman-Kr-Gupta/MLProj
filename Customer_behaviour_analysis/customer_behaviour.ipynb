```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("//content//drive//MyDrive//30daysmlchallenge//Customer_Behaviour.csv")
```

```python
df1=df.drop(['User ID','Gender'],axis='columns')
```

```python
df1.head(5)
```

```python
x=df1.iloc[:,:-1].values
x
```

```python
y=df1.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
x_train
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
age=int(input("Enter New Customer Age:"))
salary=int(input("Enter New Customer Salary:"))
newcust=[[age,salary]]
result=model.predict(sc.transform(newcust))
if(result):
  print('Customer will buy the Product')
else:
  print('Customer will not buy the Product')
```
